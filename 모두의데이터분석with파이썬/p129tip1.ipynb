```python
# p128 tip 같은 이름 다른 지역 검색 : 두 단어 띄어 쓰기 

import csv
import matplotlib.pyplot as plt

# 운영체제별 한글 폰트 설정
import os
os.name
if os.name == 'posix': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif os.name == 'nt': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정

f = open('gender202008.csv', 'r', encoding='utf-8-sig')
data = csv.reader(f)
m = []
f = []

name = input('찾고 싶은 지역 이름 : ')
for row in data :
    if name in row[0] :
        print(row[0])
f.close()

f = open('gender202008.csv', 'r', encoding='utf-8-sig')
data = csv.reader(f)
name = input('지역코드 : ')
for row in data :
    if name in row[0] :
        for i in row[3:104] :
            m.append(-int(i))
        for i in row[106:] :
            f.append(int(i))
        break
# f.reverse()

plt.style.use(['dark_background'])
plt.figure(figsize = (20,10), dpi = 300)
# plt.style.use('ggplot')
plt.title('2020.8. '+row[0]+'의 성별 인구 분포', color='orange', size='20')
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.xticks(range(-300,300,50))
plt.yticks(range(0,101,10))
plt.legend()
plt.show()
```
