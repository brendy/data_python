```python
# 4-12 지하철 시간대별 데이터 시각화하기
# 4-12-3 입력한 시간의 승차인원이 가장 많은 역

import csv

f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)

mx = 0
mx_station = ''
t = int(input('몇 시의 승차인원이 가장 많은 역이 궁금하신가요? : '))

for row in data :
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거
    del row[-1]                           # 마지막열 작업 날짜 시간 제거
    
    row[4:] = map(int, row[4:])
    a = row[4+(t-4)*2]
    if a > mx :
        mx = a
        mx_station = row[3]+'('+row[1]+')'
    
print(mx_station, mx)
```

```python
# 4-12 지하철 시간대별 데이터 시각화하기
# 4-12-4 시간대별 사람들이 가장 많이 타고 내리는 역

import csv

f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)

mx = [0] * 24      # 시간대별 최대 승차 인원 저장 리스트 초기화
mx_station = [''] * 24  # 시간대별 최대 승차 인원 역 이름 저장 리스트 초기화

for row in data :
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거
    del row[-1]                           # 마지막열 작업 날짜 시간 제거
    row[4:] = map(int, row[4:])

    for j in range(24) : 
        a = row[j*2+4]     # j와 인덱스 번호 사이의 관계식 사용
        if a > mx[j] :
            mx[j] = a
            mx_station[j] = row[3]
    
print(mx_station, mx)
```

```python
# 4-12 지하철 시간대별 데이터 시각화하기
# 4-12-4 시간대별 사람들이 가장 많이 타고 내리는 역 막대그래프

# 운영체제별 한글 폰트 설정
import os
os.name
if os.name == 'posix': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif os.name == 'nt': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정

import csv
import matplotlib.pyplot as plt

f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)

mx = [0] * 24      # 시간대별 최대 승차 인원 저장 리스트 초기화
mx_station = [''] * 24  # 시간대별 최대 승차 인원 역 이름 저장 리스트 초기화

for row in data :
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거
    del row[-1]                           # 마지막열 작업 날짜 시간 제거
    row[4:] = map(int, row[4:])

    for j in range(24) : 
        a = row[j*2+4]     # j와 인덱스 번호 사이의 관계식 사용
        if a > mx[j] :
            mx[j] = a
            mx_station[j] = row[3]+'('+str((j+4)%24)+'시)'

#plt.style.use('ggplot')
plt.style.use('dark_background')
plt.figure(figsize = (10,5), dpi=600)
plt.title('시간대별 사람들이 가장 많이 타고 내리는 역', color='orange', size='20')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
# 4-12 지하철 시간대별 데이터 시각화하기
# 4-12-5 모든 지하철역에서 시간대별 승하차 인원을 모두 더하면

import csv
import matplotlib.pyplot as plt

# 운영체제별 한글 폰트 설정
import os
os.name
if os.name == 'posix': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif os.name == 'nt': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정

f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)

s_in = [0] * 24      # 승차 인원 저장 리스트 초기화
s_out = [0] * 24     # 하차 인원 저장 리스트 초기화
t = []

for i in range(4,28) : # 시각 담기
    t.append(i%24)
    
for row in data :
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거
    del row[-1]                           # 마지막열 작업 날짜 시간 제거
    row[4:] = map(int, row[4:])

    for i in range(24) : 
        s_in[i] += row[i*2+4]
        s_out[i] += row[i*2+5]

#plt.style.use('ggplot')
plt.style.use('dark_background')
plt.figure(figsize = (10,5), dpi=600)
plt.title('지하철 시간대별 승하차 인원 추이', color='orange', size='20')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24), t)
plt.show()
```

```python
# 4-12 지하철 시간대별 데이터 시각화하기
# 하루 중 각 시간대 승차인원이 최고점에 다다른 역

import csv

f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)

mxup = 0
mxup_t = 0

mxup_station = [[] for i in range(24)]   # 2차원 배열 https://hashcode.co.kr/questions/693/파이썬에서-2차원-배열은-어떻게-쓰나요

for row in data :
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거
    del row[-1]                           # 마지막열 작업 날짜 시간 제거
    row[4:] = map(int, row[4:])
    
    for i in range(24) :
        if row[i*2+4] > mxup :
            mxup = row[i*2+4]
            mxup_t = i
    mxup_station[mxup_t].append(row[3])
    mxup = 0
    mxup_t = 0

print('*** 하루 중 각 시간대 승차인원이 최고점에 다다른 역 ***')
for i in range(24) : 
    if len(mxup_station[i]) > 0 :
        print('')
    print((i+4)%24, '시 :', mxup_station[i])
```
