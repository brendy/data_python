```python
##### 4-12 지하철 시간대별 데이터 시각화하기
    # '시간대'와 '승하차' 입력 받은 후
    # 그 시간대 승하차인원이 최고점에 다다른 역들의
    # 하루 승객 변화 그래프 그리기

import csv
import matplotlib.pyplot as plt

# 운영체제별 한글 폰트 설정
import os
os.name
if os.name == 'posix': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif os.name == 'nt': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정

f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)

t = input('알고 싶은 시간대 (0~23): ')
t = int(t)
updn = input('승차를 알고 싶으면 1, 하차를 알고 싶으면 2 : ')
updn = int(updn)

mx = 0               # 각 줄의 시간대별 최고 승객 수 임시 저장
mx_t = 0             # 각 줄의 시간대별 최고 승객 시간 임시 저장

popin = []           # 각 줄의 하루동안의 승객 수 변화 임시 저장  
result = []          # 원하는 시각이 최고점인 역의 하루동안의 승객 수 변화 저장
mx_station = [[] for i in range(24)]   # 시간대별 최고점 역 분류 리스트 2차원 배열 https://hashcode.co.kr/questions/693/파이썬에서-2차원-배열은-어떻게-쓰나요
tm = []              # 그래프 가로축에 표현할 시간 라벨
for i in range(4,28) :
    tm.append(i%24)
        
for row in data :
        
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거
    del row[-1]                           # 마지막열 작업 날짜 시간 제거
    row[4:] = map(int, row[4:])
  
    if updn == 1 :                        # 승차를 선택했을 경우
        for i in range(24) :
            if row[i*2+4] > mx :
                mx = row[i*2+4]
                mx_t = i
            popin.append(row[i*2+4])
                
    elif updn == 2 :                      # 하차를 선택했을 경우
        for i in range(24) :
            if row[i*2+5] > mx :
                mx = row[i*2+5]
                mx_t = i
            popin.append(row[i*2+5])
    
    if mx_t+4 == t :                      # 최고점 시각이 입력 시각과 일치하면
        result.append(popin)              # 해당 역의 하루동안의 승객 변화를 입력
        
    mx_station[mx_t].append(row[3])       # 해당 역 이름 저장
    mx = 0                                # 임시 변수 초기화
    mx_t = 0
    popin = []

print('')
if updn == 1 :
    print('*** 하루 중 각 시간대 승차인원이 최고점에 다다른 역 ***')
if updn == 2 :
    print('*** 하루 중 각 시간대 하차인원이 최고점에 다다른 역 ***')
print('')
print(t, '시 :', mx_station[t-4])

plt.style.use('ggplot')
plt.style.use('dark_background')
plt.figure(figsize = (10,7), dpi=600)

t = str(t)
if updn == 1 :
    plt.title(t+'시 대 승차인원이 최고점에 다다른 역의 시간대별 승차 인원 추이', color='orange', size='16')
if updn == 2 :
    plt.title(t+'시 대 하차인원이 최고점에 다다른 역의 시간대별 하차 인원 추이', color='orange', size='16')

t = int(t)
for i in range(len(mx_station[t-4])) :
    plt.plot(result[i], label=mx_station[t-4][i])
plt.legend()
plt.xticks(range(24), tm)
plt.show()
```
