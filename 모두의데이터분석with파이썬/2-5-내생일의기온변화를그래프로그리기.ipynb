```python
# p72 3)   

#    
import csv
import matplotlib.pyplot as plt

#    
import os
os.name
if os.name == 'posix': # Mac   
    plt.rc('font', family='AppleGothic')
elif os.name == 'nt': # Windows   
    plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False) #   

#  
f = open('dgutf8sig.csv', encoding='utf-8-sig')
data = csv.reader(f)
next(data)
high = []    #     
low = []      #     
year = []

for row in data :
    if row[4] != '' and row[3] != '' :   #    
        if 1975 <= int(row[0].split('.')[0]) :    # 1975  
            if row[0].split('.')[1]  == '7' and row[0].split('.')[2] == '31' :  #7 31
                high.append(float(row[4]))    #  high  
                low.append(float(row[3]))     #  low  
                year.append(float(row[0].split('.')[0]))    #  

plt.figure(figsize = (10,4))  #  10,  2 /       .
plt.title('    ')
plt.plot(year, high, 'hotpink', label=' ')
plt.plot(year, low, 'skyblue', label=' ')
plt.legend(loc = 1) # 
plt.show()
```

```python
# p73 3)     

#    
import csv
import matplotlib.pyplot as plt

#    
import os
os.name
if os.name == 'posix': # Mac   
    plt.rc('font', family='AppleGothic')
elif os.name == 'nt': # Windows   
    plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False) #   

#  
f = open('dgutf8sig.csv', encoding='utf-8-sig')
data = csv.reader(f)
next(data)

year = 1909
temp = 0
count = 0
year_list = []    #     
ave_list = []   #     

for row in data :
    if row[2] != '':   #   
        if year == int(row[0].split('.')[0]) :
            temp = float(temp) + float(row[2])
            count+=1
            
        elif year < int(row[0].split('.')[0]) :
            year_list.append(int(year))
            ave_list.append(float(temp/count))
            
            year += 1
            temp = 0
            count = 0

plt.figure(figsize = (15,4))
plt.title('    ')
plt.plot(year_list, ave_list, 'hotpink', label=' ')
plt.legend(loc = 5) # 
plt.show()
```
