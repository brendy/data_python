```python
### 지역 간 인구 비교 2 : 나이대(0~19, 20~39, 40~59, 60~79, 80~99)비율 ###

import csv
import matplotlib.pyplot as plt

# 운영체제별 한글 폰트 설정
import os
os.name
if os.name == 'posix': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif os.name == 'nt': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정

f = open('dgpop202008.csv', encoding='utf-8-sig')
data = csv.reader(f)
next(data)

age = 0
area = []
count = []
line = []
pal = ["red", "orange", "yellow", "blue", "purple"]
a = 0

for row in data :
    if '00000000)' in row[0] :    # 특정 지역 찾기 https://metagenomics.tistory.com/entry/여러번-찾기
        area.append(row[0].split(' ')[0])

        for i in range(0, 5) :
            for j in range(20*i + 3, 20*i + 23) :
                age += int(row[j].replace(",", ""))
            line.append(age)
            age = 0
                
        count.append(line)
                
        line = []

for i in range(0, len(area)) :    
    if i < 10 :
        area[i] = area[i][0:2]     # 지역 이름이 길어서 자름  https://wikidocs.net/2838
    elif i >= 10 :
        area[i] = area[i][0:3]
        
print(area)
print(count)

plt.style.use(['dark_background'])
plt.figure(figsize = (10,5), dpi=600)

plt.title('지역별 인구 비교', color='orange', size='20')
for i in range(0, len(area)) :  
    plt.bar(area[i], count[i][0]*100/sum(count[i]), color=pal[0])  
    for j in range(0, 4) : 
        a += count[i][j]
        plt.bar(area[i], count[i][j+1]*100/sum(count[i]), color=pal[j+1], bottom=a*100/sum(count[i]))
    a = 0    
```
