```python
# p66 1) 데이터에 질문하기 - 데이터 리스트에 저장하기

import csv
f = open('dgutf8sig.csv', encoding='utf-8-sig')
data = csv.reader(f)
next(data)
result = []   # 리스트 생성

for row in data :
    if row[-1] != '' :
        result.append(float(row[4])) #result에 최고 기온값 추가


print(len(result))
print(result)
```

```python
# p68 2) 데이터 시각화하기

# 파이썬 시각화 패키지 불러오기
import matplotlib.pyplot as plt
import csv

# 운영체제별 한글 폰트 설정
import os
os.name
if os.name == 'posix': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif os.name == 'nt': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정

# 본 프로그램
f = open('dgutf8sig.csv', encoding='utf-8-sig')
data = csv.reader(f)
next(data)
result = []   # 리스트 생성

for row in data :
    if row[-1] != '' :
        result.append(float(row[4])) #result에 최고 기온값 추가

plt.figure(figsize = (15,4))  # 가로로 10인치, 세로로 2인치 / 제일 위에 써야 함.
plt.title('전체 기온')
plt.plot(result, 'red', label='기온')
plt.legend(loc = 1) #범례 위치
plt.show()
```

```python
# p72 3) 날짜 데이터 추출하기

# 파이썬 시각화 패키지 불러오기
import csv
import matplotlib.pyplot as plt

# 운영체제별 한글 폰트 설정
import os
os.name
if os.name == 'posix': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif os.name == 'nt': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정

# 본 프로그램
f = open('dgutf8sig.csv', encoding='utf-8-sig')
data = csv.reader(f)
next(data)
high = []    # 최고 기온값을 저장할 리스트 생성
low = []      # 최저 기온값을 저장할 리스트 생성
year = []

for row in data :
    if row[4] != '' and row[3] != '' :   #최고 기온값과 최저 기온값이 존재한다면
        if 1975 <= int(row[0].split('.')[0]) :    # 1975년 이후 데이터라면
            if row[0].split('.')[1]  == '7' and row[0].split('.')[2] == '31' :  #7월 31일이라면
                high.append(float(row[4]))    #최고 기온값을 high 리스트에 저장
                low.append(float(row[3]))     #최저 기온값을 low 리스트에 저장
                year.append(float(row[0].split('.')[0]))    #년도 저장 

plt.figure(figsize = (10,4))  # 가로로 10인치, 세로로 2인치 / 그래프 출력 중 제일 위에 써야 함.
plt.title('내 생일의 최고 최저 기온')
plt.plot(year, high, 'hotpink', label='최고 기온')
plt.plot(year, low, 'skyblue', label='최저 기온')
plt.legend(loc = 1) #범례 위치
plt.show()
```

```python
# p73 3) 평균 기온이 상승하는 그래프 표현

# 파이썬 시각화 패키지 불러오기
import csv
import matplotlib.pyplot as plt

# 운영체제별 한글 폰트 설정
import os
os.name
if os.name == 'posix': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif os.name == 'nt': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정

# 본 프로그램
f = open('dgutf8sig.csv', encoding='utf-8-sig')
data = csv.reader(f)
next(data)

year = 1909
temp = 0
count = 0
year_list = []    # 년도 값을 저장할 리스트 생성
ave_list = []   # 평균 기온값을 저장할 리스트 생성

for row in data :
    if row[2] != '':   # 평균기온 값이 존재한다면
        if year == int(row[0].split('.')[0]) :
            temp = float(temp) + float(row[2])
            count+=1
            
        elif year < int(row[0].split('.')[0]) :
            year_list.append(int(year))
            ave_list.append(float(temp/count))
            
            year += 1
            temp = 0
            count = 0

plt.figure(figsize = (15,4))
plt.title('대구의 연 평균 기온 변화')
plt.plot(year_list, ave_list, 'hotpink', label='평균 기온')
plt.legend(loc = 5) #범례 위치
plt.show()
```
