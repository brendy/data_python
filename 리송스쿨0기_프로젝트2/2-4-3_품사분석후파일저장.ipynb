```python
# 품사별 추출해서 파일 저장하기

import pandas as pd
from khaiii import KhaiiiApi
import csv
import datetime

dt = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
df = pd.read_csv('./물체.csv', encoding='utf-8-sig')
posts = df.get('title')+" "+df.get('description')

api = KhaiiiApi()

f = open('형태소_'+str(dt)+'.csv','a', newline='', encoding='utf-8-sig')
wr = csv.writer(f)

for post in posts :
    for word in api.analyze(post) :
        print(word)
        wr.writerow([word])

f.close()
```
