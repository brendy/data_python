```python
# 4-12 지하철 시간대별 데이터 시각화하기
# 4-12-1 데이터 정제하기

import csv
f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)

for row in data :
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거

    del row[-1]     # 마지막열 작업 날짜 시간 제거
    row[4:] = map(int, row[4:])

    print(row)

```

```python
# 4-12 지하철 시간대별 데이터 시각화하기
# 4-12-2 출근 시간대 사람들이 가장 많이 타고 내리는 역

import csv
f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)
result = []

for row in data :
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거
    del row[-1]     # 마지막열 작업 날짜 시간 제거
    row[4:] = map(int, row[4:])
    result.append(row[10])
    
print(len(result))
print(result)

```

```python
# 4-12 지하철 시간대별 데이터 시각화하기
# 4-12-2 출근 시간대 사람들이 가장 많이 타고 내리는 역

import csv
import matplotlib.pyplot as plt

f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)
result = []

for row in data :
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거
    del row[-1]     # 마지막열 작업 날짜 시간 제거
    row[4:] = map(int, row[4:])
    result.append(row[10])

plt.style.use(['dark_background'])
plt.figure(figsize = (10,5), dpi=600)
plt.bar(range(len(result)), result)
plt.show
```

```python
# 4-12 지하철 시간대별 데이터 시각화하기
# 4-12-2 출근 시간대 사람들이 가장 많이 타고 내리는 역

import csv
import matplotlib.pyplot as plt

f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)
result = []

for row in data :
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거
    del row[-1]     # 마지막열 작업 날짜 시간 제거
    row[4:] = map(int, row[4:])
    result.append(row[10])

result.sort()

plt.style.use(['dark_background'])
plt.figure(figsize = (10,5), dpi=600)
plt.bar(range(len(result)), result)
plt.show
```

```python
# 4-12 지하철 시간대별 데이터 시각화하기
# 4-12-2 출근 시간대 사람들이 가장 많이 타고 내리는 역

import csv
import matplotlib.pyplot as plt

f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)
result = []

for row in data :
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거
    del row[-1]     # 마지막열 작업 날짜 시간 제거
    row[4:] = map(int, row[4:])
    result.append(sum(row[10:15:2]))

result.sort()

plt.style.use(['dark_background'])
plt.figure(figsize = (10,5), dpi=600)
plt.bar(range(len(result)), result)
plt.show
```

```python
# 4-12 지하철 시간대별 데이터 시각화하기
# 4-12-2 출근 시간대 사람들이 가장 많이 타고 내리는 역

import csv
import matplotlib.pyplot as plt

f = open('subwaytime2008.csv', encoding='utf-8-sig')
data = csv.reader(f)

next(data)
next(data)
result = []
mxup = 0
mxup_station = ''
mxdn = 0
mxdn_station = ''

for row in data :
    for i in range(len(row)) :
        row[i] = row[i].replace(",","")   # 인원 수에서 쉼표 제거
    del row[-1]                           # 마지막열 작업 날짜 시간 제거
    
    row[4:] = map(int, row[4:])

    if sum(row[10:15:2]) > mxup :
        mxup = sum(row[10:15:2])
        mxup_station = row[3]+' '+row[1]
    if sum(row[11:16:2]) > mxdn :
        mxdn = sum(row[11:16:2])
        mxdn_station = row[3]+' '+row[1]

print('가장 많이 타는 역 :', mxup_station, mxup, '명')
print('가장 많이 내리는 역 :', mxdn_station, mxdn, '명')
```
